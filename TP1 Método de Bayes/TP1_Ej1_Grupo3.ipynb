# Cálculo de probabilidades

```python
# Se guardan los datos de cada columna en un array

nacionalidad = list(datos.nacionalidad)
scones = list(datos.scones)
cerveza = list(datos.cerveza)
whiskey = list(datos.whiskey)
avena = list(datos.avena)
futbol = list(datos.futbol)
```

```python
# Cálculo de probabilidades a priori con correción de Laplace

ingleses = sum(1 for i in nacionalidad if i == "I")
p_I = (sum(1 for i in nacionalidad if i == "I")+1) / (len(datos)+2)

escoseses = len(datos) - ingleses
p_E = 1 - p_I
```

```python
# Cálculo de las probabilidades condicionales para cada valor de los atributos

# Caso inglés
p_scones_I = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "I" and scones[i] == 1) + 1) / (ingleses+2)
p_cerveza_I = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "I" and cerveza[i] == 1) + 1) / (ingleses+2)
p_whiskey_I = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "I" and whiskey[i] == 1) + 1) / (ingleses+2)
p_avena_I = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "I" and avena[i] == 1) + 1) / (ingleses+2)
p_futbol_I = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "I" and futbol[i] == 1) + 1) / (ingleses+2)

# Caso escosés
p_scones_E = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "E" and scones[i] == 1) + 1) / (escoseses+2)
p_cerveza_E = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "E" and cerveza[i] == 1) + 1) / (escoseses+2)
p_whiskey_E = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "E" and whiskey[i] == 1) + 1) / (escoseses+2)
p_avena_E = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "E" and avena[i] == 1) + 1) / (escoseses+2)
p_futbol_E = (sum(1 for i in range(len(nacionalidad)) if nacionalidad[i] == "E" and futbol[i] == 1) + 1) / (escoseses+2)
```

```python
# Cálculo de P complemento en cada caso y guardar en matrices

probs_I = [[1-p_scones_I, p_scones_I],
    [1-p_cerveza_I, p_cerveza_I],
    [1-p_whiskey_I, p_whiskey_I],
    [1-p_avena_I, p_avena_I],
    [1-p_futbol_I, p_futbol_I]]

probs_E = [[1-p_scones_E, p_scones_E],
    [1-p_cerveza_E, p_cerveza_E],
    [1-p_whiskey_E, p_whiskey_E],
    [1-p_avena_E, p_avena_E],
    [1-p_futbol_E, p_futbol_E]]
```

```python
# Clasificador ingenuo de Bayes

# x = (scones, cerveza, whiskey, avena, futbol)
# Ejemplos de la consigna:
x_1 = (1,0,1,1,0)
x_2 = (0,1,1,0,1)

def clasificador_Bayes(x, p_I, p_E, probs_I, probs_E):
  v_NB = [] # Se calcula el v_NB para cada clase, no las probabilidades ya que no las pide

  v_NB.append(p_I*probs_I[0][x[0]]*probs_I[1][x[1]]*probs_I[2][x[2]]*probs_I[3][x[3]]*probs_I[4][x[4]]) # Caso inglés
  v_NB.append(p_E*probs_E[0][x[0]]*probs_E[1][x[1]]*probs_E[2][x[2]]*probs_E[3][x[3]]*probs_E[4][x[4]]) # Caso escosés

  # probs_I[0][x[0]] --> p_scones en la matriz de probs_I; el valor del x para ese atributo define si es la p de que le gusten o no

  if v_NB[0] > v_NB[1]: # Si el primer v_NB es mayor, es inglés
    clasificacion = "inglés"
  elif v_NB[0] < v_NB[1]: # Si el segundo v_NB es mayor, es escosés
    clasificacion = "escosés"
  else: # Si por alguna razón no entra a ningún caso anterior, la clasificación es indefinida (evita que se rompa el código)
    clasificacion = "indefinido"

  return clasificacion, v_NB[0], v_NB[1]

# Probamos con x_1
clasificacion_1, v_NB_I_1, v_NB_E_1 = clasificador_Bayes(x_1, p_I, p_E, probs_I, probs_E)

print("Las preferencias del ejemplo 1 corresponden a un", clasificacion_1)
print("v_NB_I =", v_NB_I_1)
print("v_NB_E =", v_NB_E_1)
print("\n")

# Probamos con x_2
clasificacion_2, v_NB_I_2, v_NB_E_2 = clasificador_Bayes(x_2, p_I, p_E, probs_I, probs_E)

print("Las preferencias del ejemplo 2 corresponden a un", clasificacion_2)
print("v_NB_I =", v_NB_I_2)
print("v_NB_E =", v_NB_E_2)
```
