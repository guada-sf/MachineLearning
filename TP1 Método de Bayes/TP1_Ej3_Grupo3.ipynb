# Discretización de variables

```python
# GRE >= 500 --> 1
# GRE < 500 --> 0

for i in range(len(datos)): # Mira cada dato y si el GRE es >= 500 sobreescribe el valor con 1, si no lo sobreescribe con 0
  if datos.gre[i] >= 500:
    datos.at[i, "gre"] = 1
  else:
    datos.at[i, "gre"] = 0
```

```python
# GPA >= 3 --> 1
# GPA < 3 --> 0

for i in range(len(datos)):  # Mira cada dato y si el GPA es >= 3 sobreescribe el valor con 1, si no lo sobreescribe con 0
  if datos.gpa[i] >= 3:
    datos.at[i, "gpa"] = 1
  else:
    datos.at[i, "gpa"] = 0

datos['gpa'] = datos['gpa'].astype(int) # Pasamos todos los valores de GPA int (antes estaban en float)
```

# Cálculo de probabilidades

## Probabilidad del rango

```python
# Creamos un DF para que quede como tabla
# Los indices son los rangos (1 a 4)
# Se inicializan en 1 por la corrección de Laplace

p_rango = pd.DataFrame(1, index=['1','2','3','4'], columns=['true', 'false'])

# Cálculo de frecuencias para cada caso
for i in range(len(datos)):
  if datos.at[i, "rank"] == 1:
    p_rango.at['1', "true"] += 1
  elif datos.at[i, "rank"] == 2:
    p_rango.at['2', "true"] += 1
  elif datos.at[i, "rank"] == 3:
    p_rango.at['3', "true"] += 1
  elif datos.at[i, "rank"] == 4:
    p_rango.at['4', "true"] += 1

total = len(datos)

for i in range(1,5):
  p_rango.at[str(i), "true"] /= (total + 4) # Corrección de Laplace

for i in range(1,5):
  p_rango.at[str(i), "false"] -= p_rango.at[str(i), "true"] # Se calcula como 1 - p_true (complemento)
```

## Probabilidad del GRE

```python
# Se crea un DF para que quede como tabla
# Los indices son los rangos
indices = ['rank_1','rank_2','rank_3','rank_4']

p_gre = pd.DataFrame(1, index=indices, columns=['GRE_0', 'GRE_1']) # Datos iniciales = 1 por la corrección de Laplace

# Cálculo de frecuencias para cada caso
for i in range(len(datos)):
  if datos.at[i, "rank"] == 1 and datos.at[i, "gre"] == 0:
    p_gre.at['rank_1', 'GRE_0'] += 1
  elif datos.at[i, "rank"] == 1 and datos.at[i, "gre"] == 1:
    p_gre.at['rank_1', 'GRE_1'] += 1
  elif datos.at[i, "rank"] == 2 and datos.at[i, "gre"] == 0:
    p_gre.at['rank_2', 'GRE_0'] += 1
  elif datos.at[i, "rank"] == 2 and datos.at[i, "gre"] == 1:
    p_gre.at['rank_2', 'GRE_1'] += 1
  elif datos.at[i, "rank"] == 3 and datos.at[i, "gre"] == 0:
    p_gre.at['rank_3', 'GRE_0'] += 1
  elif datos.at[i, "rank"] == 3 and datos.at[i, "gre"] == 1:
    p_gre.at['rank_3', 'GRE_1'] += 1
  elif datos.at[i, "rank"] == 4 and datos.at[i, "gre"] == 0:
    p_gre.at['rank_4', 'GRE_0'] += 1
  elif datos.at[i, "rank"] == 4 and datos.at[i, "gre"] == 1:
    p_gre.at['rank_4', 'GRE_1'] += 1

for i in indices:
  suma = p_gre.sum(1)[i] # Suma de cada fila --> total de cada rango
  p_gre.at[i, "GRE_0"] /= (suma - 2 + 2) 
  p_gre.at[i, "GRE_1"] /= (suma - 2 + 2)
  # Se cancela la corrección de Laplace ya que se suma dos veces 1 en los datos iniciales del DF
```

# Item a

```python
# P (admit = 0 | rank = 1)

# numerador = P (rank = 1, admit = 0, GPA, GRE) = sumar los 4 valores de la columna admit_0 para rank 1

# denominador = P (rank = 1)

# 1 termino num = p(rank=1, admit=0, GPA=0, GRE=0) * p(GPA=0|rank=1) * p(GRE=0|rank=1) * p(rank=1)
# 2 termino num = p(rank=1, admit=0, GPA=1, GRE=0) * p(GPA=1|rank=1) * p(GRE=0|rank=1) * p(rank=1)
# 3 termino num = p(rank=1, admit=0, GPA=0, GRE=1) * p(GPA=0|rank=1) * p(GRE=1|rank=1) * p(rank=1)
# 4 termino num = p(rank=1, admit=0, GPA=1, GRE=1) * p(GPA=1|rank=1) * p(GRE=1|rank=1) * p(rank=1)
```

```python
numerador = 0

for i in range(4): # Sabemos que son las primeras 4 probabilidades del DF p_admit

  if i%2 == 0: # para los elementos del indice = 0 e indice = 2 que tienen GPA = 0
    p_cond_gpa = p_gpa.at['rank_1', 'GPA_0'] # p(GPA=0|rank=1)
    
  else: # para los elementos del indice = 1 e indice = 3 que tienen GPA = 1
    p_cond_gpa = p_gpa.at['rank_1', 'GPA_1'] # p(GPA=1|rank=1)

  if i == 0 or i == 1: # tienen GRE = 0
    p_cond_gre = p_gre.at['rank_1', 'GRE_0'] # p(GRE=0|rank=1)

  elif i == 2 or i == 3: # tienen GRE = 1
    p_cond_gre = p_gre.at['rank_1', 'GRE_1'] # p(GRE=0|rank=1)

  termino = p_admit.at[i, 'admit_0'] * p_cond_gpa * p_cond_gre * p_rango.at['1', 'true']

  numerador += termino

prob_a = numerador / p_rango.at['1', 'true'] # P (admit = 0 | rank = 1)

print(prob_a)
```

# Item c

El aprendizaje en las redes bayesianas es un proceso que consiste en dos pasos. Por un lado el aprendizaje de la estructura, esto es, la estimación de la estructura del grafo. Por otro lado, el aprendizaje paramétrico, donde a partir de datos observados se estiman las probabilidades a priori y condicionales. 

En este ejercicio en particular, se lleva a cabo el aprendizaje paramétrico, aplicando la corrección de Laplace, lo que permite modelar y hacer inferencias probabilísticas sobre las relaciones entre las variables rango, GRE, GPA y admisión de los estudiantes.
